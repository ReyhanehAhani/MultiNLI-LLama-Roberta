```python
from transformers import TrainingArguments, Trainer
```

## Pre-Processing and Loading

```python
from transformers import RobertaForSequenceClassification, RobertaTokenizer
from datasets import load_dataset
```

```python
from datasets import load_dataset

dataset = load_dataset("nyu-mll/multi_nli")
```

```python
train_len = len(dataset['train'])
```

```python
dataset['train'] = dataset['train'].shuffle(seed=42).filter(lambda example, idx: idx < train_len * 0.25, with_indices=True)
```

```python
dataset
```

```python
dataset['train'][0]
```

```python
model = RobertaForSequenceClassification.from_pretrained('FacebookAI/roberta-large', num_labels=3)
tokenizer = RobertaTokenizer.from_pretrained('FacebookAI/roberta-large')
```

```python
def tokenize_func(examples):
    return tokenizer(examples['premise'], examples['hypothesis'], truncation=True)

tokenized_dataset = dataset.map(tokenize_func, batched=True)
```

```python
from transformers import DataCollatorWithPadding

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
import evaluate

accuracy = evaluate.load("accuracy")
```

```python
import numpy as np

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1)
    return accuracy.compute(predictions=predictions, references=labels)
```

```python
from peft import LoraConfig, get_peft_model, TaskType

config = LoraConfig(
    task_type=TaskType.SEQ_CLS,
    r=32,
    lora_alpha=16,
    target_modules=["query", "value"],
    lora_dropout=0.1,
)

model = get_peft_model(model, config)
model.print_trainable_parameters()
```

```python
model
```

## Training Process

```python
import torch
```

```python
with torch.no_grad():
    torch.cuda.empty_cache()
```

```python
training_args = TrainingArguments(
    output_dir = './result',
    save_steps=0.5,
    learning_rate=2e-5,
    num_train_epochs=1,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=4,
    warmup_steps=0.1,
    weight_decay=0.01,
    report_to='none',
    eval_strategy="steps",
    eval_steps=1/4,
    overwrite_output_dir=True,
    fp16=True,
)


trainer = Trainer(
    model = model,
    args = training_args,
    train_dataset = tokenized_dataset['train'],
    eval_dataset = tokenized_dataset['validation_matched'],
    compute_metrics=compute_metrics,
    data_collator=data_collator,
    tokenizer=tokenizer,
)
```

```python
trainer.train()
```

```python
trainer.evaluate()
```

```python
trainer.evaluate(eval_dataset=tokenized_dataset['validation_mismatched'])
```
